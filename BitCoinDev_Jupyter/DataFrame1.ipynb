```python
from autotrading.db.mongodb import mongodb_handler
mongodb = mongodb_handler.MongoDBHandler("local","coiner","price_info")

query = {
    "cointype":"ADA",
    "Currency_Trade":"Bithumb",
    "year" : 2018,
    "month" : 7
}
result = mongodb.find_items(query)

result = list(result)
```

## 기본적인 함수를 이용한 데이터 가공
* 데이터 프레임에서 기본적으로 제공하는 함수들 중 중요한 몇가지를 이용해서 데이터를 가공해 보자

### 1. df의 모든 칼럼에 1.05가 곱해진 df_2를 확인해보자

```python
df = pd.DataFrame(result,columns=['amount', 'price','timestamp'])
df['amount'] = df['amount'].astype(float)
df['price'] = df['price'].astype(float)
df['timestamp'] = df['timestamp'].astype(float)
df_2 = df * 1.05
df_2
```

### 2. df 와 df_2를 이용하여 서로 다른 데이터프레임 간의 데이터 가공에 대해 알아보자
* 두 개의 데이터프레임을 더해 보자. + 연산자나 데이터프레임의 add메서드를 이용하면 된다.
* 연산자 외에도 기본적인 산술 연산자에 대응하는 메서드가 있다.
> * add() :  + 
> * sub() :  -
> * mul() :  *
> * div() : /

```python
df + df_2
df.add(df_2)
```

### 3. 비교연산자 또한 같으며 결과가 boolean 타입으로 반환된다.
 * gt 외에도 여러가지 비교 연산자를 제공한다.
  >  * eq() : ==
  >  * ne() : !=
  >  * it() : <
  >  * gt() : >
  >  * le() : <=
  >  * ge() : >=

```python
df > df_2
df.gt(df_2)
```

### Boolean Reductions에 대해 알아보자. Reduction는 한글로 축소, 삭감이라는 의미이며, 여러가지 값을 하나의 값으로 축소시킨다는 의미 정도로 이해 할 수 있다.  Reductions 메서드는 empty(), any(), all(), bool()이 있다.




### 4. 먼저, 180보다 큰 값이 있는지 비교 한 후 any()를 호출하면 각 칼럼에 대해 불린 값이 반환된다. any()는 칼럼 값 중 180보다 큰 값이 하나 이상 인 경우 True를 반환하며 0개인 경우 False를 반환된다.

```python
(df > 180).any()
```

### 5. all()인 경우 모든 칼럽의 값이 180보다 큰 경우  True를 반환하고, 아니면 False를 반환한다.

```python
(df > 181).all()
```

### 6. empty는 데이터프레임의 모든 칼럼값이 비어 있는지 여부를 반환한다. 아무 값이 없는 데이터프레임을 생선하면 True를 반환한다.

```python
df.empty
```

### 7. 아무값이없는 데이터프레임을 생성한다면 True를 반환한다.
