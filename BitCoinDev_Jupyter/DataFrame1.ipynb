## 기본적인 함수를 이용한 데이터 가공
* 데이터 프레임에서 기본적으로 제공하는 함수들 중 중요한 몇가지를 이용해서 데이터를 가공해 보자

### 1. df의 모든 칼럼에 1.05가 곱해진 df_2를 확인해보자

```python
df = pd.DataFrame(result,columns=['amount', 'price','timestamp'])
df['amount'] = df['amount'].astype(float)
df['price'] = df['price'].astype(float)
df['timestamp'] = df['timestamp'].astype(float)
df_2 = df * 1.05
df_2
```

### 2. df 와 df_2를 이용하여 서로 다른 데이터프레임 간의 데이터 가공에 대해 알아보자
* 두 개의 데이터프레임을 더해 보자. + 연산자나 데이터프레임의 add메서드를 이용하면 된다.
* 연산자 외에도 기본적인 산술 연산자에 대응하는 메서드가 있다.
> * add() :  + 
> * sub() :  -
> * mul() :  *
> * div() : /

```python
df + df_2
df.add(df_2)
```

### 3. 비교연산자 또한 같으며 결과가 boolean 타입으로 반환된다.
 * gt 외에도 여러가지 비교 연산자를 제공한다.
  >  * eq() : ==
  >  * ne() : !=
  >  * it() : <
  >  * gt() : >
  >  * le() : <=
  >  * ge() : >=

```python
df > df_2
df.gt(df_2)
```

### Boolean Reductions에 대해 알아보자. Reduction는 한글로 축소, 삭감이라는 의미이며, 여러가지 값을 하나의 값으로 축소시킨다는 의미 정도로 이해 할 수 있다.  Reductions 메서드는 empty(), any(), all(), bool()이 있다.




### 4. 먼저, 180보다 큰 값이 있는지 비교 한 후 any()를 호출하면 각 칼럼에 대해 불린 값이 반환된다. any()는 칼럼 값 중 180보다 큰 값이 하나 이상 인 경우 True를 반환하며 0개인 경우 False를 반환된다.

```python
(df > 180).any()
```

### 5. all()인 경우 모든 칼럽의 값이 180보다 큰 경우  True를 반환하고, 아니면 False를 반환한다.

```python
(df > 181).all()
```

### 6. empty는 데이터프레임의 모든 칼럼값이 비어 있는지 여부를 반환한다. 아무 값이 없는 데이터프레임을 생선하면 True를 반환한다.

```python
df.empty
```

### 7. 아무 값이없는 데이터프레임을 생성한다면 True를 반환한다.

```python
pd.DataFrame(columns=list('ABC')).empty
```

### 8. Reductions 메서드를 이용하면 데이터프레임 간의 연산 결과를 손쉽게 확인 할 수 있다.

## * 통계(Descriptive Statistics) 관련 메서드

#### 다양한 통계 함수를 제공하며 공식 문서에서 설명하는 함수의 종류는 다음과 같다.
* Count : NA값을 제외한 값의 수를 반환
* describe : 시리즈 혹은 데이터프레임의 각 열에 대한 요약 통계
* min,max : 최소, 최대값
* armin, argmax : 최소, 최대값을 갖고 있는 색인 위치 반환
* idxmin, idxmanx : 최소, 최대값 갖고 있는 색인의 값 반환
* quantile : 0부터 1까지의 분위수 계산
* sum : 합
* mean : 평균
* median : 주우이값
* mad : 평균값에서 절대 평균 편차
* var : 표본 분산
* std : 표본 정규분산
* skew : 표본 비대칭도
* kurt : 표본 첨도
* cumsum : 누적 합
* cummin, cummax : 누적 최소값, 누적 최대값
* cumprod :  누적 곱
* diff : 1차 산술차(시계열 데이터 사용시 유용)
* pct_change : 퍼센트 변화율 계산

### 2. 반복(iteration)문 또한 쓸 수 있는데, 데이터 프레임에서는 iterrows(), itertyples(), iteritems() 메서드를 제공한다.
 * pandas 공식 문서에서는 데이터프레임의 반복문 수행은 일반적으로 느리다고 경고하고 있음.
 * 반복문을 수행하면서 각 행에 대한 값을 절대 변경하지 말라고 경고하고 있음.
 * 반복문을 수행하지 않고도 각 행에 함수를 적용하거나 값을 변경할 수 있는 기능들을 제공 하고 있기 떄문에, 특별한 경우가 아닌 경우 반복문 사용을 권고하지 않음.

#### iterrows()는 idx, row라는 2개의 항목을 반환한다. idx는 Index를 반환하며, row는 각 행의 시리즈(Series)를 반환한다.

```python
for idx,row in df.iterrows():
    print(idx)
    print(row)
```

#### itertuples()는 각 행의 namedtuple을 반환한다.
* iteritems()는 키-값 쌍(key-value pairs)으로 값을 반환한다.

```python
for row in df.itertuples():
    print(row)
```

#### item은 칼럼의 이름을 반환하며 frame은 item에 해당하는 칼럼의 시리즈를 반환한다.

```python
for item, frame in df.iteritems():
    print(item)
    print(frame)
```
